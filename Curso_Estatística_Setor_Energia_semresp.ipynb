```python
frequencia_brasil = pd.crosstab(final_df['Fonte Geração'],
                         final_df['UF']
                        )
frequencia_brasil.rename(index = fontes, inplace = True)
frequencia_brasil.rename(columns = uf, inplace = True)
frequencia_brasil
```

> Sintaxe da função:
> 1. frequencia_brasil - variável armazenária a tabela
> 2. pd.crosstab(final_df['Fonte Geração'], final_df['UF']) - esta parte utiliza a função pd.crosstab() para criar uma tabela de contingência (crosstab) que contém a contagem de ocorrências das diferentes fontes de geração em relação aos estados (UF) presentes no DataFrame final_df. Os valores únicos na coluna "Fonte Geração" se tornam as linhas da tabela, e os valores únicos na coluna "UF" se tornam as colunas.

```python
percentual_brasil = pd.crosstab(final_df['Fonte Geração'],
                         final_df['UF'],
                         normalize = True
                        ) * 100
percentual_brasil.rename(index = fontes, inplace = True)
percentual_brasil.rename(columns = uf, inplace = True)
percentual_brasil
```
